```python
#Neste notebook serão removidas as colunas "created_at", "updated_at" e "average_distance", pois são colunas que não serão utilizadas.

#preprocessing is done

import pandas as pd

df_jb_opp_workstyle = pd.read_csv('Dados - bettha/job_opportunity_workstyle_consolidates.csv')
```

```python
df_jb_opp_workstyle.drop('average_distance', axis = 1, inplace = True)
```

```python
df_jb_opp_workstyle
```

```python
df_jb_opp_workstyle['created_at'] = pd.to_datetime(df_jb_opp_workstyle['created_at'])
```

```python
df_jb_opp_workstyle['updated_at'] = pd.to_datetime(df_jb_opp_workstyle['updated_at'])
```

```python
df_jb_opp_workstyle['updated_at'] = df_jb_opp_workstyle['updated_at'].dt.strftime('%d/%m/%Y')
```

```python
df_jb_opp_workstyle.drop('created_at', axis = 1, inplace = True)
```
