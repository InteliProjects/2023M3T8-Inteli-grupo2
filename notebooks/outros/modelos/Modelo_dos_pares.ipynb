***USERS***

Este código lida com a leitura dos dados do arquivo CSV chamado 'superfit.csv' usando a biblioteca pandas e a criação de objetos da classe User com base nessas informações.

Cada objeto User representa um usuário e armazena várias características relacionadas ao "superfit", como 'superfit_dis', 'superfit_sin', 'superfit_cur', entre outros. O código percorre as linhas do arquivo CSV, cria objetos User para cada linha e os armazena em uma lista chamada users.


No geral, o código é uma etapa inicial para processar e representar os dados de usuários em objetos Python para análise posterior.

```python
#Impoertando a biblioteca do pandas e abreviando como pd para poder
#tratar os dados e manusealos
import pandas as pd

#Carregamento da tabela chamada 'superfit.csv'
df = pd.read_csv('superfit.csv')

#Classe de usuários
# Aqui, uma classe chamada User é definida. Esta classe representa os usuários e suas características relacionadas ao "superfit".
# Ela tem um construtor __init__ que aceita vários atributos, como id, superfit_dis, dentre outros para poder criar os objetos
class User:
    def __init__(self, id, superfit_dis, superfit_sin, superfit_cur, superfit_int, superfit_eng, superfit_res):
        self.id = id
        self.superfit_dis = superfit_dis
        self.superfit_sin = superfit_sin
        self.superfit_cur = superfit_cur
        self.superfit_int = superfit_int
        self.superfit_eng = superfit_eng
        self.superfit_res = superfit_res

# Aqui temos uma lista vazia para poder armazenar os objetos do tipo User
# e como o código vai percorrendo todas as linhas, vai armazenando todos os users nesta lista vazia
users = []

#Criando os objetos da classe User com base na database
# rodando a função temos um loop que percorre pelas características listadas
for index, row in df.iterrows():
    user = User(row['id'], row['superfit_dis'], row['superfit_sin'], row['superfit_cur'], row['superfit_int'], row['superfit_eng'], row['superfit_res'])
    users.append(user)

#for user in users:
    #print(f"id: {user.id}, superfit_dis: {user.superfit_dis}, superfit_sin: {user.superfit_sin}, superfit_cur: {user.superfit_cur}, superfit_int: {user.superfit_int}, superfit_eng: {user.superfit_eng}, superfit_res: {user.superfit_res}")
```

***JOBS***

Este código lida com a leitura dos dados de um arquivo CSV chamado 'jobs.csv' usando a biblioteca pandas e a criação de objetos da classe Job com base nessas informações.

Cada objeto Job representa uma vaga de emprego e armazena várias características relacionadas ao "superfit", como 'superfit_dis', 'superfit_sin', 'superfit_cur', entre outros. O código percorre as linhas do arquivo CSV, cria objetos Job para cada vaga de emprego com base nas informações da linha e os armazena em uma lista chamada jobs.

Seguindo o comentário do código acima que difere só em relação a basse de dados usada, no geral, o código é uma etapa inicial para processar e representar os dados de vagas de emprego em objetos Python para uma análise em conjunto com os dados obtidos do códifo acima.






```python
#Importando a base de dados jobs para poder usar e tratar os dados
df = pd.read_csv('jobs.csv')

#Classe de vagas
# Este trecho, uma classe chamada Job é definida. Esta classe representa vagas de emprego e armazena
# as características relacionadas ao "superfit", como 'superfit_dis', 'superfit_sin', 'superfit_cur', dentre os outros.
# Ela possui um construtor __init__ que aceita essas características como argumentos e cria objetos do tipo vaga de emprego com base neles.
class Job:
    def __init__(self, id, superfit_dis, superfit_sin, superfit_cur, superfit_int, superfit_eng, superfit_res):
        self.id = id
        self.superfit_dis = superfit_dis
        self.superfit_sin = superfit_sin
        self.superfit_cur = superfit_cur
        self.superfit_int = superfit_int
        self.superfit_eng = superfit_eng
        self.superfit_res = superfit_res

# Aqui temos uma lista vazia para poder armazenar os objetos do tipo Jobs
# e como o código vai percorrendo todas as linhas, vai armazenando todos os jobs nesta lista vazia
jobs = []

#Criando os objetos da classe Job com base na databse
# rodando a função temos um loop que percorre pelas características listadas
for index, row in df.iterrows():
    job = Job(row['id'], row['score_dis'], row['score_sin'], row['score_cur'], row['score_int'], row['score_eng'], row['score_res'])
    jobs.append(job)

#for job in jobs:
   # print(f"id: {job.id}, superfit_dis: {job.superfit_dis}, superfit_sin: {job.superfit_sin}, superfit_cur: {job.superfit_cur}, superfit_int: {job.superfit_int}, superfit_eng: {job.superfit_eng}, superfit_res: {job.superfit_res}")
```

***MODELO***

Este código implementa um sistema de recomendação de vagas de emprego com base na compatibilidade das características "superfit" entre os usuários e as vagas.

E com isso, faz os cálculos de similaridade, cálculo de compatibilidade percentual, recomendação das melhores vagas, recomendação de vagas para todos os usuários e o tratamento de caso sem recomendações.

Em resumo, o código utiliza cálculos de similaridade e compatibilidade para recomendar vagas de emprego aos usuários, ajudando a identificar as vagas mais adequadas com base em suas características "superfit".

```python

#Cálculo da similaridade entre um usuário e uma vaga
# Criando a função para aplicar no modelo fazendo os calculos necessários para
# juntar os pares que vão se formar
def calculate_similarity(user, job):
    similarity_score = (
        abs(user.superfit_dis/90 - job.superfit_dis/58) +
        abs(user.superfit_sin/90 - job.superfit_sin/58) +
        abs(user.superfit_cur/90 - job.superfit_cur/58) +
        abs(user.superfit_int/90 - job.superfit_int/58) +
        abs(user.superfit_eng/90 - job.superfit_eng/58) +
        abs(user.superfit_res/90 - job.superfit_res/58)
    )

    return similarity_score

#Cálculo da compatibilidade percentual entre o usuário e a vaga
# Outra função para usar os dados de User e de job fazer essa determinação dos
# pares de forma percentual
def calculate_compatibility(user, job):

    similarity_score = calculate_similarity(user, job)
    compatibility_percentage = (1 - similarity_score)*100

    return compatibility_percentage

#Recomendação das três melhores vagas para um usuário
#Com base, nos dados do candidato, temos a recomendação
#de 3 empresas para poder mostrar o possivel "match" dele.
def recommend_top_jobs(user, jobs, num_recommendations=3):
    top_jobs = []

#Função percorrendo as informações do candidato para saber quais são as empresas
# que seria recomendada para ele.
    for job in jobs:
        similarity_score = calculate_similarity(user, job)

        if len(top_jobs) < num_recommendations:
            top_jobs.append((job, similarity_score))
        else:
            max_score_job, max_score = max(top_jobs, key=lambda x: x[1])

            if similarity_score < max_score:
                top_jobs.remove((max_score_job, max_score))
                top_jobs.append((job, similarity_score))

    top_jobs.sort(key=lambda x: x[1])

    return top_jobs

#Recomendação de vagas para todos os usuários
for user in users:
    top_jobs = recommend_top_jobs(user, jobs, num_recommendations=3)

    if top_jobs:
        print(f"Opções mais compatíveis para o usuário {user.id}:")
        for i, (job, similarity_score) in enumerate(top_jobs, start=1):
            compatibility_percentage = calculate_compatibility(user, job)
            print(f"{i}. {job.id} (Compatibilidade: {compatibility_percentage:.2f}%)")
    else:
        print(f"No job recommendations available for User {user.id}")
```

Este código funciona para calcular a compatibilidade e recomendar as três melhores vagas de emprego para o usuário (User1 em questão) com base nas características "superfit" especificadas para esse usuário. Ele calcula a similaridade entre User1 e cada vaga, classifica as vagas com base na similaridade e imprime as três melhores opções com suas respectivas porcentagens de compatibilidade.

```python
# Função para calcular a similaridade entre o User1 e uma vaga
def calculate_similarity_user1(job):
    # Calcula a similaridade somando as diferenças normalizadas entre as características "superfit" do User1 e da vaga
    similarity_score_user1 = (
        abs(0.655172 - job.superfit_dis/58) +
        abs(0.620690 - job.superfit_sin/58) +
        abs(0.482759 - job.superfit_cur/58) +
        abs(0.758621 - job.superfit_int/58) +
        abs(0.586207 - job.superfit_eng/58) +
        abs(0.724138 - job.superfit_res/58)
    )

    return similarity_score_user1

# Função para calcular a compatibilidade percentual entre o User1 e uma vaga
def calculate_compatibility_user1(job):
    # Calcula a compatibilidade percentual com base na similaridade
    similarity_score_user1 = calculate_similarity_user1(job)
    compatibility_percentage_user1 = (1 - similarity_score_user1) * 100

    return compatibility_percentage_user1

# Função para recomendar as melhores vagas para o User1
def recommend_top_jobs_user1(jobs, num_recommendations=3):
    top_jobs = []

    for job in jobs:
        similarity_score_user1 = calculate_similarity_user1(job)

        if len(top_jobs) < num_recommendations:
            top_jobs.append((job, similarity_score_user1))
        else:
            max_score_job, max_score = max(top_jobs, key=lambda x: x[1])

            if similarity_score_user1 < max_score:
                top_jobs.remove((max_score_job, max_score))
                top_jobs.append((job, similarity_score_user1))

    top_jobs.sort(key=lambda x: x[1])

    return top_jobs

# Chamada da função para recomendar as melhores vagas para o User1
top_jobs = recommend_top_jobs_user1(jobs, num_recommendations=3)

# Imprime as opções mais compatíveis para o usuário User1
print(f"Opções mais compatíveis para o usuário:")
for i, (job, similarity_score) in enumerate(top_jobs, start=1):
    compatibility_percentage_user1 = calculate_compatibility_user1(job)
    print(f"{i}. {job.id} (Compatibilidade: {compatibility_percentage_user1:.2f}%)")
```

***MÉTRICAS E TUNING DE HIPERPARÂMETROS***

Ja este código, com o uso dos hiperparâmetros, realiza uma busca em grade com validação cruzada para determinar os melhores hiperparâmetros para um modelo de recomendação. Ele usa o erro quadrático médio (MSE) como métrica de avaliação. A função evaluate_model é usada para calcular o MSE médio para diferentes hiperparâmetros, e o resultado é impresso na tela, mostrando os melhores hiperparâmetros e o melhor MSE obtido durante a busca em grade.

```python
# Importação de bibliotecas necessárias
from sklearn.model_selection import GridSearchCV  # Para busca de hiperparâmetros
from sklearn.metrics import mean_squared_error  # Para calcular o erro quadrático médio

# Função para avaliar o modelo de recomendação
def evaluate_model(users, jobs, num_recommendations=3, hyperparameters=None):
    mse_scores = []  # Lista para armazenar os valores do erro quadrático médio (MSE)

    # Itera por cada usuário
    for user in users:
        # Recomenda as melhores vagas para o usuário com base nos hiperparâmetros fornecidos
        top_jobs = recommend_top_jobs(user, jobs, num_recommendations=num_recommendations, hyperparameters=hyperparameters)

        if top_jobs:
            # Calcula os escores reais de compatibilidade para as vagas recomendadas
            actual_scores = [calculate_compatibility(user, job) for job, _ in top_jobs]

            # Exemplo: Calcula escores previstos com base na posição das vagas
            predicted_scores = [100 - i * 10 for i in range(len(top_jobs))]

            # Calcula o erro quadrático médio (MSE) entre os escores reais e previstos
            mse = mean_squared_error(actual_scores, predicted_scores)
            mse_scores.append(mse)

    if mse_scores:
        # Calcula a média dos MSEs para todos os usuários
        average_mse = sum(mse_scores) / len(mse_scores)
        return average_mse
    else:
        return None

# Definindo uma grade de hiperparâmetros para busca
param_grid = {
    'num_recommendations': [2, 3, 4, 5],
}

# Realiza busca em grade com validação cruzada para determinar os melhores hiperparâmetros
grid_search = GridSearchCV(estimator=None, param_grid=param_grid, scoring='neg_mean_squared_error', cv=3)
grid_search.fit(users, jobs)  # Ajusta o modelo usando dados de usuários e vagas

# Obtém os melhores hiperparâmetros e o melhor MSE a partir da busca em grade
best_hyperparameters = grid_search.best_params_
best_mse = -grid_search.best_score_  # O valor é negativo, pois o GridSearchCV maximiza o score

# Imprime os melhores hiperparâmetros e o melhor MSE
print("Melhores hiperparâmetros:", best_hyperparameters)
print("Melhor mínimo erro quadrático:", best_mse)
```
