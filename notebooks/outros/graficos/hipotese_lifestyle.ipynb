```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('prep_df_user_lifestyle.csv')

# Conta a frequncia de cada valor nico na coluna 'key'
key_counts = df['key'].value_counts()

# Ordena os valores pela frequncia, em ordem decrescente
key_counts = key_counts.sort_values(ascending=False)

# Cria o grfico de barras
plt.figure(figsize=(10, 6))
plt.bar(key_counts.index, key_counts.values)

# Adiciona rtulos e ttulos
plt.xlabel('Keys')
plt.ylabel('Quantidade')
plt.title('Frequncia de Keys')

# Adiciona a contagem acima das barras
for i, v in enumerate(key_counts.values):
    plt.text(i, v + 1, str(v), ha='center', va='bottom')

plt.xticks(rotation=45, ha='right')

# Exibe o grfico
plt.show()
```
