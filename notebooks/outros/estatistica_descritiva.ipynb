#Estatística descritiva dos dados
Este notebook descreve como o grupo Ebettha obteve os dados de estatística descritiva do conjunto de dados do projeto.

Na célula abaixo é fornecido o código para se obter os principais dados descritivo para as tabelas. É importante substituir a variável "life_path" com o caminho para o arquivo que estamos querendo analisar.

```python
import pandas as pd

# Carregar o arquivo CSV em um DataFrame.
file_path = 'prep_df_jb_opp_workstyle.csv' #Substituir essa variável com o caminho para o arquivo no qual queremos analisar.
df = pd.read_csv(file_path)

# Calcular estatística descritiva das colunas da tabelas
df.describe()
```

O código acima gerou várias informações relevantes, que são:


*   count: Número de linhas da tabela;
*   mean: Média aritmética dos valores da coluna;
*   std: Desvio padrão dos valores da coluna;
*   min e max: Valores mínimo e máximo, respectivamente, da coluna;
*   25%, 50% e 75%: Valores dos percentis 25, 50 e 75, respectivamente, de cada coluna.

Um outro dado que é válido termos na nossa análise, mas que não é apresentado com o comando acima é a moda de cada coluna, isto é, o valor que mais se repete em cada uma das colunas. Isso pode ser feito com o comando abaixo:

```python
#Calcular a moda
df.mode()
```
