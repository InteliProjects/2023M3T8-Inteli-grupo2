```python
#Criando uma verificação para comparar todos os resultados únicos de cada um dos resultados possíveis do teste de lifestyle

valores_procurados = ['ls_explorer_generalist', 'ls_order_hybrid',
                      'ls_explorer_specialist', 'ls_change_specialist',
                      'ls_tireless_hybrid', 'ls_change_generalist',
                      'ls_tireless_generalist', 'ls_tireless_specialist',
                      'ls_change_hybrid', 'ls_explorer_hybrid', 'ls_order_specialist',
                      'ls_order_generalist', 'ls_classic_hybrid',
                      'ls_classic_specialist', 'ls_classic_generalist']

# Crie um dicionário onde as chaves são os valores procurados e os valores são as linhas correspondentes
linhas_por_condicao = {}

for valor in valores_procurados:
    filtro = df_user_lifestyle_inteli['key'] == valor
    df_condicao = df_user_lifestyle_inteli[filtro]

    # Verifique se o DataFrame não está vazio antes de adicionar ao dicionário
    if not df_condicao.empty:
        linhas_por_condicao[valor] = df_condicao.iloc[0]  # Escolha a primeira linha correspondente

# Agora você tem um dicionário onde cada chave é uma condição e o valor é a linha correspondente (a primeira linha encontrada)

type(linhas_por_condicao)

df_resultado = pd.DataFrame.from_dict(linhas_por_condicao, orient='index')

df_resultado

#Esse dataframe é divido em duas partes: estilo de trabalho e características
```
